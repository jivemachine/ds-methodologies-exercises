```python
from scipy import stats
import pandas as pd
import numpy as np

import sklearn.preprocessing
from sklearn.model_selection import train_test_split

import pydataset

df = pydataset.data('sat.act')

df.head()
```

```python
# 1. make the thing
scaler = sklearn.preprocessing.MinMaxScaler()
# 2. Fit the thing
scaler.fit(df[['age']])
# 3. Use the thing
scaler.transform(df[['age']])
```

```python
df.head()
```

```python
df_scaled = df.copy()

scaler = sklearn.preprocessing.MinMaxScaler()

scaler.fit(df[['age']])
df_scaled['age'] = scaler.transform(df[['age']])

scaler.fit(df[['ACT']])
df_scaled['ACT'] = scaler.transform(df[['ACT']])

scaler.fit(df[['SATV']])
df_scaled['SATV'] = scaler.transform(df[['SATV']])
```

```python
df_scaled.head()
```

```python
scaler = sklearn.preprocessing.MinMaxScaler()

df_scaled = df.copy()

scaler.fit(df[['age', 'ACT', 'SATV', 'SATQ']])

df_scaled[['age', 'ACT', 'SATV', 'SATQ']] = scaler.transform(df[['age', 'ACT', 'SATV', 'SATQ']])

df_scaled.head()
                                                                
```

scaling with train test dataset

```python
df = pydataset.data('sat.act')
df.head()


#do the train test split

train, test = train_test_split(df, random_state=123, train_size=.8)

```

```python
scaler = sklearn.preprocessing.MinMaxScaler()

scaler.fit(train)

train_scaled = scaler.transform(train)

test_scaled = scaler.transform(test)

train_scaled
```

```python
scaler = sklearn.preprocessing.MinMaxScaler()

scaler.fit(train)

# turn the resulting numpy array into a dataframe
train_scaled = pd.DataFrame(scaler.transform(train), columns=train.columns, index=train.index )

train_scaled
```
