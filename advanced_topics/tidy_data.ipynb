```python
sales = pd.read_csv('untidy-data/sales.csv')
sales
```

```python
year_and_measure_df = sales_melt['year_&_measure'].str.split(' ', expand=True)
year_and_measure_df.columns = ['year', 'measure']
year_and_measure_df.head()
```

```python
sales2 = pd.concat([sales_melt, year_and_measure_df], axis=1).drop(columns='year_&_measure')
sales2.head()
```

```python
sales_tidy = sales2.pivot_table(index=['Product', 'year'], columns='measure', values='value')
sales_tidy.head()
```

```python
sales_tidy.columns.name = ''
sales_tidy.reset_index(inplace=True)
sales_tidy
```

# Exercise 1
## Attendance Data

Load the attendance.csv file and calculate an attendnace percentage for each student. One half day is worth 50% of a full day, and 10 tardies is equal to one absence.

You should end up with something like this:

`name
Billy    0.5250
Jane     0.6875
John     0.9125
Sally    0.7625
Name: grade, dtype: float64`


```python
df = pd.read_csv('untidy-data/attendance.csv')
df.head()
```

```python
df['name'] = df['Unnamed: 0']
df = df.drop(columns='Unnamed: 0')
df.head()
```

```python
df = df.melt(id_vars='name')
df
```

```python
df = df.drop(columns=['variable'])
```

```python
df.head()
```

```python
attendance_percentage = {
    'P' : 1,
    'T' : .9,
    'H' : 0.5,
    'A' : 0
}

df['attendance_percentage'] = df.value.map(attendance_percentage)
df.head(3)
```

```python
df = df.drop(columns='value')
```

```python
df.groupby('name')['attendance_percentage'].mean()
```

# Exercise 3
## Cake Recipes

    a. Read the cake_recipes.csv data. This data set contains cake tastiness scores for combinations of different recipes, oven rack positions, and oven temperatures.

```python
df = pd.read_csv('untidy-data/cake_recipes.csv')
```

    b. Tidy the data as necessary.

```python
df.head()
```

```python
cut = df["recipe:position"].str.split(':').reset_index()
cut.head()
df['position'] = cut['recipe:position'].str[1]
df['recipe'] = cut['recipe:position'].str[-2]
df = df.drop(columns='recipe:position')
df.head()
```

```python
df_melt = df.melt(id_vars=['recipe', 'position'], var_name='temperature', value_name='rating')
df_melt.head()
```

    c. Which recipe, on average, is the best? recipe b

```python
df_melt.groupby('recipe').rating.mean()
```

#### B is the best cake on average

    d. Which oven temperature, on average, produces the best results? 275

```python
df_melt.groupby('temperature').rating.mean()
```

#### 275 on avg is producing the best results

    e. Which combination of recipe, rack position, and temperature gives the best result? recipe b, bottom rack, 300 degrees

```python
df_melt.groupby(['recipe', 'position','temperature']).rating.mean().sort_values(ascending=False)
```

#### recipe b, bottom postion and 300 temp cake has thhe highest score
